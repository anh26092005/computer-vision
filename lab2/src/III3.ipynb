```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Đọc ảnh
img = cv2.imread('../input/lab_image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Các bộ lọc
mean = cv2.blur(img_rgb, (15,15))
gaussian = cv2.GaussianBlur(img_rgb, (15,15), 0)
median = cv2.medianBlur(img_rgb, 15)
bilateral = cv2.bilateralFilter(img_rgb, 15, 150, 150)
```

```python
# Hiển thị
titles = ["Original", "Mean", "Gaussian", "Median", "Bilateral"]
images = [img_rgb, mean, gaussian, median, bilateral]

plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### Nhận xét:
- Mean: làm mờ mạnh, mất chi tiết
- Gaussian: mượt hơn Mean, tự nhiên hơn
- Median: giữ cạnh tốt hơn, giảm chi tiết nhẹ
- Bilateral: giữ biên tốt nhất, gần giống ảnh gốc
