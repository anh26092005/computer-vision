```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Đọc ảnh
img = cv2.imread('../input/lab_image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Thêm nhiễu (để test)
noise = np.random.normal(0, 60, img_rgb.shape).astype(np.int16)
noisy_img = np.clip(img_rgb.astype(np.int16) + noise, 0, 255).astype(np.uint8)
```

```python
# ===== 2. MEDIAN FILTER =====
median = cv2.medianBlur(noisy_img, 9)
# ===== 3. BILATERAL FILTER =====
bilateral = cv2.bilateralFilter(noisy_img, 15, 150, 150)
```

```python
titles = ["Original", "Noisy (Gaussian)", "Median", "Bilateral"]
images = [img_rgb, noisy_img, median, bilateral]

plt.figure(figsize=(10,6))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
